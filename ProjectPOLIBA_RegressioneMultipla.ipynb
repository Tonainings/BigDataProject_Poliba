```python
import numpy as np

def compute_multiple_cost(X, Y, W, b):                                                    # creazione di una funzione "compute cost"

  m = X.shape[0];                                                              # dichiaro che la variabile m è uguale alla dimensione degli elementi di x

  cost = 0                                                                       # dichiaro che la variabile cost (la sola sommatoria della cost_function) sia uguale a 0
  for i in range(m):
    X_i = X[i,:]
    f_Wb = np.dot(W,X_i) + b                                                          # ciclo for, con i = 1 : m (i che varia da 1 a m) RICORDA: m è la dimensione degli elementi di x!!
    #f_wb = w*x[i] + b
    err_i = f_Wb - Y[i]                                                           # x(i) ,, f_wb è la funzione di una generica retta
    cost = cost + (err_i)**2                                               # incrementa il valore di cost con la somma del valore di cost della precedente iterazione e di ciò che è tra parentesi, dove:
                                                                                 #f_wb è la retta presa in considerazione e y[i] è la distanza (altezza) del punto con la retta.
                                                                                 #cost è stato incrementato, seguirà una nuova iterazione.
    total_cost = (1/(2*m))*cost                                                    # concluso il ciclo for, total_cost sarà la cost_function completa
  return total_cost                                                              # return chiude la funzione compute_cost e mi conserva la variabile locale total_cost per il resto del codice. Rendendola globale



total_cost = compute_multiple_cost(X,Y,W,b)

print(total_cost)
```
